**Aula: Lidando com Tabelas de dados (*data frames*)**

As Tabelas de Dados (data frames) são estruturas semelhantes a uma matriz, tratando-se de um arranjo retangular de valores, ou seja, não é mais um “vetor unidimensional” e suas dimensões percorrem duas direções: (número de linhas) x (número de colunas). A principal diferença entre as duas estruturas diz respeito ao tipo de dados que recebem (https://rstudio-education.github.io/hopr/):

image.png


**DICAS:** 

1) O data frame não aceita o mesmo nome para duas variáveis ​​diferentes

2) todos os elementos (colunas) são vetores e têm o mesmo comprimento (linhas)

3) os vetores são arranjados no formato de data frame com a função *data.frame()*

**Definindo o tipo de variável**

Nem sempre o padrão de texto (character) ou numérico (integer, double) atende às nossas necessidades de análises. Assim, nos depararemos com situações onde é necessário definir o tipo de dado de forma mais clara. Fazemos isso com as funções "as". Por exemplo, para definir um vetor como fator, usamos a função as.factor. Outras opções são as.numeric, as.character, as.logical, as.Date etc. A conversão para data merece uma atenção especial, dado que pode ser obtida em vários formatos, que devem ser indicados na linguagem R de acordo com seus símbolos:

image.png

**DICAS**

1) Em R, as datas são armazenadas como o número de dias desde 01/01/1970, com valores negativos para datas anteriores (Hora Unix) 

2) A função factor() permite tansformar caracteres em variáveis categóriicas ordinais ao setar o argumento order = True e definir os níveis no argimento levels

3) A função format() permite extrair o valor de dia, mês ou ano de um campo de data. Ex: format(x,%d) extrai o dia do campo de data x
