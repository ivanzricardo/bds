**Instalando e chamando pacotes**

Pacotes são conjuntos de funções pré-estabelecidas que auxiliam variadas tarefas em R. Ex: o pacote “epitools” possui funções conhecidas na área de cálculos epidemiológicos.  

Até agora estávamos apenas lidando com operadores internos e funções de pacotes pré-instalados (built-in) na linguagem R. A maioria das funções discutidas até agora faz parte do pacote *base*

*Instalando pacotes*

Existem duas forma principais de instalação de pacotes:

**Opção 1**
Os pacotes validados e disponibilizados pela rede CRAN, podem ser instalados com a função install.packages()
image.png

**Opção 2)** Versões ou pacotes ainda não disponibilizados na rede CRAN podem ser instalados diretamente do github com a instalação prévia do pacote devtools

image.png

*Chamando pacotes*

Não é necessário instalar os pacotes toda vez que utilizar seu ambiente. Contudo, sempre será necessário chamá-lo antes do uso. Fazemos isso com a função library():

image.png



**DICAS**

1) Mesmo pacotes que já estão na rede CRAN pode estar com versões defasadas. Por isso, é bem comum a instalação alternativa

2) GitHub é uma plataforma de hospedagem de código-fonte e arquivos com controle de versão (https://github.com/) 

3) O microdatasus é um pacote do R pdoerosíssimo para manipulação de bancos de dados do SUS. Saiba mais em https://github.com/rfsaldanha/microdatasus

**Importando um arquivo CSV**

A maneira mais prática de importarmos um arquivo csv em R é com uso das funções read.csv ou read.csv2 do pacote base e com as funções read_csv ou read_csv2 do pacote tidyverse:
image.png

image.png

**DICAS**

1) É possível definir o padrão de importação de cada coluna com uso do argumento col_types na função read_csv ou read_csv2

2) As funções head() e tail() permitem visualizar os 5 primeiros e 5 últimos registros do data frame, respectivamente

3) Nas definições de importação, variáveis de data podem ser importadas como tal ao definir sua coluna com o formato de data disponível na fonte de dados. Ex: um caracter aramzenado como "25-06-98" pode ser importando como data definindo sua coluna com col_date(format = "%d-%m%y")
